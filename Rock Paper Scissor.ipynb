```python
from random import randint

options = ["Rock", "Paper", "Scissors"]

computer = options[randint(0,2)]

player = ""

Your_points = 0
Ties = 0
Computer_Points = 0

def display(Your_points, Ties, Computer_Points):
    print("----------------------------------------------------------------------")
    print("Your Points: ",Your_points," Ties: ",Ties," Computer's points: ",Computer_Points)
    print("----------------------------------------------------------------------")
```

```python
print("Welcome, Good Luck\nMake your First choice wisely :);)\nType 'exit' to stop playing...")
while player != "Exit":
    player = input("Rock, Paper, Scissors? ")
    player = player.title()
    if player == computer:
        print("Tie!!!")
        Ties += 1
        display(Your_points, Ties, Computer_Points)
    elif player == "Rock":
        if computer == "Paper":
            print("You lose!!! ", computer, " Wraps ", player)
            Computer_Points += 1
            display(Your_points, Ties, Computer_Points)
        else:
            print("You win!!! ", player, " Breaks ", computer)
            Your_points += 1
            display(Your_points, Ties, Computer_Points)
    elif player == "Paper":
        if computer == "Scissors":
            print("You lose!!! ", computer, " Cuts ", player)
            Computer_Points += 1
            display(Your_points, Ties, Computer_Points)
        else:
            print("You win!!! ", player, " Wraps ", computer)
            Your_points += 1
            display(Your_points, Ties, Computer_Points)
    elif player == "Scissors":
        if computer == "Rock":
            print("You lose!!! ", computer, " Breaks ", player)
            Computer_Points += 1
            display(Your_points, Ties, Computer_Points)
        else:
            print("You win!!! ", player, " Cuts ", computer)
            Your_points += 1
            display(Your_points, Ties, Computer_Points)
    elif player == "Exit":
      display(Your_points, Ties, Computer_Points)
      if Your_points > Computer_Points:
          print("Congratulations, You played really well!!")
      else:
          print("Better luck next time :(")
      print("Thank you for playing")
    else:
      print("That's not a valid play. Check your spelling!")
    computer = options[randint(0,2)]
```
